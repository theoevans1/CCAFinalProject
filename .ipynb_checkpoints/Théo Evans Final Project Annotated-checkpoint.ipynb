Projecting words onto a gender dimension

```python
def normalize(vector):
    normalized_vector = vector / np.linalg.norm(vector)
    return normalized_vector

def dimension(model, positives, negatives):
    diff = sum([normalize(model[x]) for x in positives]) - sum([normalize(model[y]) for y in negatives])
    return diff
```

```python
Gender = dimension(fanficW2V2, ['man','boy','he','him','guy'], ['woman', 'girl','she','her','lady'])
```

```python
CharacterTypes = ["leader","companion","hero","victim","lover", "sweetie","villain","monster","boyfriend", "girlfriend","date"]
Adjectives = ["evil","good","kind","cruel","brave","coward","mysterious","wise","stupid","foolish"]
```

```python
def makeDF(model, word_list):
    g = []
    s = []
    for word in word_list:
        g.append(sklearn.metrics.pairwise.cosine_similarity(fanficW2V2[word].reshape(1,-1), Gender.reshape(1,-1))[0][0])
    df = pandas.DataFrame({'gender': g}, index = word_list)
    return df
```

```python
CharacterTypedf = makeDF(fanficW2V2, CharacterTypes) 
Adjectivedf = makeDF(fanficW2V2, Adjectives)
```

```python
def Coloring(Series):
    x = Series.values
    y = x-x.min()
    z = y/y.max()
    c = list(plt.cm.rainbow(z))
    return c

def PlotDimension(ax,df, dim):
    ax.set_frame_on(False)
    ax.set_title(dim, fontsize = 20)
    colors = Coloring(df[dim])
    for i, word in enumerate(df.index):
        ax.annotate(word, (0, df[dim][i]), color = colors[i], alpha = 0.6, fontsize = 12)
    MaxY = df[dim].max()
    MinY = df[dim].min()
    plt.ylim(MinY,MaxY)
    plt.yticks(())
    plt.xticks(())
```

```python
fig = plt.figure(figsize = (12,4))
ax1 = fig.add_subplot(131)
PlotDimension(ax1, CharacterTypedf, 'gender')
```

```python
def spacy_pos(word_list):
    tags = []
    doc = nlp(word_list.lower())
    for w in doc:
        tags.append((w.text, w.tag_))
    return tags
```

Now I'll merge these word lists into one larger list of adjectives

```python
def mergeDict(dict1, dict2):
    dict3 = {**dict1, **dict2}
    for key, value in dict3.items():
        if key in dict1 and key in dict2:
            dict3[key] = value + dict1[key]
    return dict3
```

```python
person_adjectives = mergeDict(mergeDict(mergeDict(man_adj, woman_adj), mergeDict(boy_adj, girl_adj)),person_adj)
sortedperson_adjectives = sorted(person_adjectives.items(), key = lambda x: x[1], reverse = True)
sortedperson_adjectives[:100]
```

```python
characteristic_adjectives = mergeDict(mergeDict(mergeDict(face_adj, voice_adj), mergeDict(hand_adj, body_adj)), mergeDict(smile_adj, skin_adj))
sortedcharacteristic_adjectives = sorted(characteristic_adjectives.items(), key = lambda x: x[1], reverse = True)
sortedcharacteristic_adjectives[:100]
```

```python
adjectives = mergeDict(mergeDict(person_adjectives, characteristic_adjectives), mind_adj)
sortedadjectives = sorted(adjectives.items(), key = lambda x: x[1], reverse = True)
sortedadjectives[:200]
```

```python
def pairwise_metric_average(metric, array):
    p = permutations(range(array[0,:].size),2)
    m = [metric(array[:,x[0]], array[:,x[1]]) for x in p]
    clean_m = [c for c in m if not math.isnan(c)]
    return reduce(lambda a, b: a + b, clean_m)/len(clean_m)    
 
pairwise_metric_average(cohens_kappa, anno.annotations)
```

```python
def makeMMWDF(model, word_list):
    g = []
    for word in word_list:
        g.append(sklearn.metrics.pairwise.cosine_similarity(MMW2V[word].reshape(1,-1), MMWoman.reshape(1,-1))[0][0])
    df = pandas.DataFrame({'Women in M/M Stories': g}, index = word_list)
    return df
```
