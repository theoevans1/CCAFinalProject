```python
supernaturalcategories_df = supernatural_df[(supernatural_df.category == "F/M")|(supernatural_df.category == "M/M")|(supernatural_df.category == "F/F")]
```

```python
holdBackFraction = .2
train_data_df, test_data_df = lucem_illud.trainTestSplit(supernaturalcategories_df, holdBackFraction=holdBackFraction)
```

```python
TFVectorizer = sklearn.feature_extraction.text.TfidfVectorizer(max_df=100, min_df=2, stop_words='english', norm='l2')
TFVects = TFVectorizer.fit_transform(train_data_df['body'])
```

```python
train_data_df['vect'] = [np.array(v).flatten() for v in TFVects.todense()]
```

```python
TFVects_test = TFVectorizer.transform(test_data_df['body'])
test_data_df['vect'] = [np.array(v).flatten() for v in TFVects_test.todense()]
```

```python
naiveBayes = sklearn.naive_bayes.BernoulliNB()
naiveBayes.fit(np.stack(train_data_df['vect'], axis=0), train_data_df['category'])
```

```python
print("Training:")
print(naiveBayes.score(np.stack(train_data_df['vect'], axis=0), train_data_df['category']))
print("Testing:")
print(naiveBayes.score(np.stack(test_data_df['vect'], axis=0), test_data_df['category']))
```

```python
clf_nn = sklearn.neural_network.MLPClassifier()
clf_nn.fit(np.stack(train_data_df['vect'], axis=0), train_data_df['category'])
```

```python
sklearn.metrics.accuracy_score(test_data_df['category'],clf_nn.predict(np.stack(test_data_df['vect'], axis = 0)))
```

```python
clf_svm = sklearn.svm.SVC(kernel='linear', probability = False)
#probability = True is slower but  lets you call predict_proba()
clf_svm.fit(np.stack(train_data_df['vect'], axis=0), train_data_df['category'])
```

```python
sklearn.metrics.accuracy_score(test_data_df['category'],clf_svm.predict(np.stack(test_data_df['vect'], axis = 0)))
```

```python
clf_tree = sklearn.tree.DecisionTreeClassifier(max_depth=4,random_state=0)
clf_tree.fit(np.stack(train_data_df['vect'], axis =0), train_data_df['category'])
```

```python
sklearn.metrics.accuracy_score(test_data_df['category'],clf_tree.predict(np.stack(test_data_df['vect'], axis = 0)))
```

```python
MM, FF, FM = naiveBayes.feature_log_prob_

words_dict = {
    'MM' : [],
    'MM_log_prob' : [],
    'FF' : [],
    'FF_log_prob' : [],
    'FM' : [],
    'FM_log_prob' : []
}

for i, prob in sorted(enumerate(MM), key = lambda x:x[1], reverse=True)[:30]:
    words_dict['MM'].append(TFVectorizer.get_feature_names()[i])
    words_dict['MM_log_prob'].append(prob)

for i, prob in sorted(enumerate(FF), key = lambda x:x[1], reverse=True)[:30]:
    words_dict['FF'].append(TFVectorizer.get_feature_names()[i])
    words_dict['FF_log_prob'].append(prob)
    
for i, prob in sorted(enumerate(FM), key = lambda x:x[1], reverse=True)[:30]:
    words_dict['FM'].append(TFVectorizer.get_feature_names()[i])
    words_dict['FM_log_prob'].append(prob)
    
pandas.DataFrame(words_dict)
```

```python
tree = sklearn.tree.DecisionTreeClassifier(max_depth=10) #Create an instance of our decision tree classifier.

bag = sklearn.ensemble.BaggingClassifier(tree, n_estimators=100, max_samples=0.8, random_state=1) #Each tree uses up to 80% of the data
```

```python
bag.fit(np.stack(train_data_df['vect'], axis =0), train_data_df['category']) #Fit the bagged classifier
```

```python
sklearn.metrics.accuracy_score(test_data_df['category'],bag.predict(np.stack(test_data_df['vect'], axis = 0)))
```
